# Task

## Continuous-continuous visualizations

### Subtask:
Generate code for line plot and scatter plot to visualize the relationship between continuous variables.


**Reasoning**:
Generate a line plot to visualize the trend of `listening_time` over `songs_played_per_day` and a scatter plot to explore the relationship between `listening_time` and `ads_listened_per_week`.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='songs_played_per_day', y='listening_time')
plt.title('Listening Time vs Songs Played Per Day')
plt.xlabel('Songs Played Per Day')
plt.ylabel('Listening Time')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='ads_listened_per_week', y='listening_time')
plt.title('Listening Time vs Ads Listened Per Week')
plt.xlabel('Ads Listened Per Week')
plt.ylabel('Listening Time')
plt.show()
```

## Categorical-categorical visualizations

### Subtask:
Generate code for dodged and stacked countplots to visualize the relationship between categorical variables.


**Reasoning**:
Generate code for dodged and stacked countplots to visualize the relationship between categorical variables as per the instructions.



```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='subscription_type', hue='is_churned')
plt.title('Churn Rate by Subscription Type (Dodged)')
plt.xlabel('Subscription Type')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='device_type', hue='is_churned', dodge=False)
plt.title('Churn Rate by Device Type (Stacked)')
plt.xlabel('Device Type')
plt.ylabel('Count')
plt.show()
```

## Categorical-continuous visualizations

### Subtask:
Generate code for boxplot and bar plot to visualize the relationship between categorical and continuous variables.


**Reasoning**:
Create a boxplot and a bar plot to visualize the relationship between categorical and continuous variables as per the instructions.



```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='subscription_type', y='listening_time')
plt.title('Listening Time Distribution by Subscription Type')
plt.xlabel('Subscription Type')
plt.ylabel('Listening Time')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='device_type', y='ads_listened_per_week')
plt.title('Average Ads Listened Per Week by Device Type')
plt.xlabel('Device Type')
plt.ylabel('Average Ads Listened Per Week')
plt.show()
```

## Subplots

### Subtask:
Generate code to combine multiple plots into subplots.


**Reasoning**:
Create a figure and subplots and then plot a scatter plot and a boxplot on the respective subplots.



```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

sns.scatterplot(data=df, x='ads_listened_per_week', y='listening_time', ax=axes[0])
axes[0].set_title('Listening Time vs Ads Listened Per Week')
axes[0].set_xlabel('Ads Listened Per Week')
axes[0].set_ylabel('Listening Time')

sns.boxplot(data=df, x='subscription_type', y='listening_time', ax=axes[1])
axes[1].set_title('Listening Time Distribution by Subscription Type')
axes[1].set_xlabel('Subscription Type')
axes[1].set_ylabel('Listening Time')

plt.tight_layout()
plt.show()
```

## Styling and labeling

### Subtask:
Add appropriate styling and labels to the plots for better readability.


**Reasoning**:
Apply styling and labeling enhancements to one of the previously created plots to improve readability and aesthetics.



```python
plt.figure(figsize=(12, 7))
sns.countplot(data=df, x='subscription_type', hue='is_churned', palette='viridis')
plt.title('Churn Rate by Subscription Type (Dodged)', fontsize=16)
plt.xlabel('Subscription Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Churned', labels=['No', 'Yes'])
plt.show()
```

**Reasoning**:
Apply styling and labeling enhancements to another previously created plot to further demonstrate the application of styling and labeling to different plot types.



```python
plt.figure(figsize=(12, 7))
sns.boxplot(data=df, x='subscription_type', y='listening_time', palette='plasma')
plt.title('Listening Time Distribution by Subscription Type', fontsize=16)
plt.xlabel('Subscription Type', fontsize=12)
plt.ylabel('Listening Time (minutes)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Reasoning**:
Apply styling and labeling enhancements to a scatter plot to demonstrate styling on continuous-continuous visualizations and showcase how to combine styling with subplots as previously generated.



```python
fig, axes = plt.subplots(1, 2, figsize=(15, 7))

sns.scatterplot(data=df, x='ads_listened_per_week', y='listening_time', ax=axes[0], alpha=0.6, color='teal')
axes[0].set_title('Listening Time vs Ads Listened Per Week', fontsize=14)
axes[0].set_xlabel('Ads Listened Per Week', fontsize=11)
axes[0].set_ylabel('Listening Time (minutes)', fontsize=11)
axes[0].grid(True, linestyle='--', alpha=0.5)

sns.boxplot(data=df, x='subscription_type', y='listening_time', ax=axes[1], palette='plasma')
axes[1].set_title('Listening Time Distribution by Subscription Type', fontsize=14)
axes[1].set_xlabel('Subscription Type', fontsize=11)
axes[1].set_ylabel('Listening Time (minutes)', fontsize=11)
axes[1].grid(axis='y', linestyle='--', alpha=0.5)


plt.tight_layout()
plt.show()
```
